```python
# price_df.to_csv("test/price/daily/test_daily_price_total.csv", index=False)

price_df = pd.read_csv("test/price/daily/test_daily_price_total.csv")
price_df
```

```python
price_df['Reversion_Returns'] = price_df['Close'].pct_change(5)
```

```python
price_df = price_df.dropna()
```

```python
price_df
```

```python
price_df['Reversion_Returns_Abs'] = price_df['Reversion_Returns'].abs()
price_df['Reversion_Flag'] = price_df['Reversion_Returns'].apply(lambda x: 1 if x >= 0 else -1)
```

```python
price_df
```

```python
price_df['Reversion_Signals'] = price_df['Reversion_Returns_Abs'].apply(lambda x: 1 if x >= 0.1 else 0)
```

```python
price_df
```

```python
price_df['Reversion_Signals_Returns'] = price_df['Reversion_Signals'] * price_df['Reversion_Returns']
```

```python
price_df['Reversion_Signals'] = price_df['Reversion_Signals'] * price_df['Reversion_Flag']
```

```python
price_df['Reversion_Signals'].value_counts()
```

```python
price_df.to_csv("test_daily_price_total_reversion_preprocessed.csv", index=False)
```
