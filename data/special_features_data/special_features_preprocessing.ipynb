```python
import pandas as pd
import numpy as np
```

```python
avg_confirmation_time_df = pd.read_csv('./special_features_data/avg-confirmation-time.csv')
market_cap_df = pd.read_csv('./special_features_data/market-cap.csv')
mempool_count_df = pd.read_csv('./special_features_data/mempool-count.csv')
mempool_growth_df = pd.read_csv('./special_features_data/mempool-growth.csv')
mempool_size_df = pd.read_csv('./special_features_data/mempool-size.csv')
n_payments_per_block_df = pd.read_csv('./special_features_data/n-payments-per-block.csv')
n_payments_df = pd.read_csv('./special_features_data/n-payments.csv')
total_bitcoins_df = pd.read_csv('./special_features_data/total-bitcoins.csv')
transactions_per_second_df = pd.read_csv('./special_features_data/transactions-per-second.csv')
utxo_count_df = pd.read_csv('./special_features_data/utxo-count.csv')
```

```python
# 각 DataFrame의 'timestamp' 컬럼을 datetime으로 변환하고 인덱스로 설정
dataframes = [
    avg_confirmation_time_df,
    market_cap_df,
    mempool_count_df,
    mempool_growth_df,
    mempool_size_df,
    n_payments_per_block_df,
    n_payments_df,
    total_bitcoins_df,
    transactions_per_second_df,
    utxo_count_df
]

for df in dataframes:
    df['timestamp'] = pd.to_datetime(df['timestamp'])
    df.set_index('timestamp', inplace=True)
```

```python
# 일별 평균 계산 (하루 단위로 리샘플링)
daily_avg = avg_confirmation_time_df.resample('D').mean()
daily_avg.index = daily_avg.index.date
daily_avg.to_csv('./special_features_data/avg-confirmation-time_preprocessed.csv')
daily_avg
```

```python
# 일별 평균 계산 (하루 단위로 리샘플링)
daily_avg = market_cap_df.resample('D').last()
daily_avg.index = daily_avg.index.date
daily_avg
daily_avg.to_csv('./special_features_data/market-cap_preprocessed.csv')
```
