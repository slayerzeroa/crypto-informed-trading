```python
coin_dominance = pd.read_csv('./data/coin_dominance/coin_dominance.csv')
coin_dominance['timestamp'] = coin_dominance['timestamp'].apply(lambda x: x[:11])
coin_dominance = coin_dominance.set_index('timestamp')
coin_dominance = coin_dominance[(coin_dominance.index >= '2018-01-01')&(coin_dominance.index <= '2024-09-30')]
coin_dominance
```

```python
import pandas as pd

# 기본 describe()
summary = coin_dominance.describe()

# First와 Last 값 계산
first = coin_dominance.iloc[0]
last = coin_dominance.iloc[-1]

# summary에 추가
summary.loc["first"] = first
summary.loc["last"] = last

# 보기 좋게 정렬 (optional: 인덱스 순서를 재정렬)
summary = summary.loc[
    ['count', 'mean', 'std', 'min', '25%', '50%', '75%', 'max', 'first', 'last']
]

print(summary)
```
